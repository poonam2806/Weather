```python
import requests

# API endpoint for the hourly weather forecast of London
API_ENDPOINT = "https://samples.openweathermap.org/data/2.5/forecast/hourly"

def get_weather_data(date):
    # Make an API call to get the weather data for the given date
    params = {"q": "London,us", "appid": "b6907d289e10d714a6e88b30761fae22"}
    response = requests.get(API_ENDPOINT, params=params)

    if response.status_code == 200:
        data = response.json()
        for item in data["list"]:
            if item["dt_txt"].split()[0] == date:
                return item["main"]["temp"]
        print("No weather data found for the input date.")
        return None
    else:
        print(f"Failed to fetch weather data. Status code: {response.status_code}")
        return None

def get_wind_speed_data(date):
    # Make an API call to get the wind speed data for the given date
    params = {"q": "London,us", "appid": "b6907d289e10d714a6e88b30761fae22"}
    response = requests.get(API_ENDPOINT, params=params)

    if response.status_code == 200:
        data = response.json()
        for item in data["list"]:
            if item["dt_txt"].split()[0] == date:
                return item["wind"]["speed"]
        print("No wind speed data found for the input date.")
        return None
    else:
        print(f"Failed to fetch wind speed data. Status code: {response.status_code}")
        return None

def get_pressure_data(date):
    # Make an API call to get the pressure data for the given date
    params = {"q": "London,us", "appid": "b6907d289e10d714a6e88b30761fae22"}
    response = requests.get(API_ENDPOINT, params=params)

    if response.status_code == 200:
        data = response.json()
        for item in data["list"]:
            if item["dt_txt"].split()[0] == date:
                return item["main"]["pressure"]
        print("No pressure data found for the input date.")
        return None
    else:
        print(f"Failed to fetch pressure data. Status code: {response.status_code}")
        return None

def main():
    while True:
        print("1. Get Weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")

        choice = int(input("Enter your choice:"))
        if choice == 1:
            date = input("Enter the date (YYYY-MM-DD):")
            weather = get_weather_data(date)
            if weather:
                print(f"The temperature on {date} is {weather}°C")
            else:
                print("Failed to retrieve weather data.")

        elif choice == 2:
            date = input("Enter the date (YYYY-MM-DD):")
            wind_speed = get_wind_speed_data(date)
            if wind_speed:
                print(f"The wind speed on {date} is {wind_speed} km/h")
            else:
                print("Failed to retrieve wind speed data.")

        elif choice == 3:
            date = input("Enter the date (YYYY-MM-DD):")
            pressure = get_pressure_data(date)
            if pressure:
                print(f"The pressure on {date} is {pressure} hPa")
            else:
                print("Failed to retrieve pressure data.")

        elif choice == 0:
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
